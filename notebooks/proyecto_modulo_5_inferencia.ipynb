## Problema de investigación

El presente estudio busca analizar los factores que influyen en los hábitos saludables de jóvenes universitarios, específicamente en dimensiones relacionadas con sueño, alimentación, actividad física y bienestar general. La finalidad es identificar patrones estadísticos relevantes que permitan comprender mejor el comportamiento de la población objetivo y generar recomendaciones útiles para futuras políticas de bienestar estudiantil.

## Pregunta de investigación

¿Existen diferencias y relaciones estadísticamente relevantes entre las horas de sueño, la actividad física, la alimentación y el bienestar percibido en jóvenes universitarios?

## Hipótesis principal

- **Hipótesis nula (H0):** No existe diferencia significativa en las horas de sueño promedio entre estudiantes con actividad física suficiente y estudiantes con actividad física insuficiente.
- **Hipótesis alternativa (H1):** Sí existe diferencia significativa en las horas de sueño promedio entre estudiantes con actividad física suficiente y estudiantes con actividad física insuficiente.

## Hipótesis secundaria

- **Hipótesis nula (H0):** La proporción de estudiantes con sueño insuficiente es igual o menor al 50%.
- **Hipótesis alternativa (H1):** La proporción de estudiantes con sueño insuficiente es mayor al 50%.

## Lección 2 - Probabilidad y estadística

### Espacio muestral y eventos de interés

En este estudio se trabajará con una población simulada de estudiantes universitarios y una muestra estratificada por año universitario. El espacio muestral estará constituido por los estudiantes seleccionados en la muestra.

Se definirán los siguientes eventos aleatorios:

- **A:** el estudiante presenta actividad física suficiente (150 minutos o más por semana).
- **B:** el estudiante presenta sueño insuficiente (menos de 7 horas de sueño por noche).
- **C:** el estudiante reporta una autopercepción de salud favorable (Buena o Muy buena).

A partir de estos eventos se calcularán probabilidades simples, conjuntas, uniones y complementarias.

### Diseño muestral

Se utilizará un **muestreo estratificado por año universitario**, debido a que este criterio permite representar de manera más equilibrada a estudiantes de distintos niveles académicos. Este tipo de muestreo resulta pertinente porque los hábitos de estudio, sueño, actividad física y estrés pueden variar según el avance en la trayectoria universitaria.

Primero se simulará una población de estudiantes, y posteriormente se extraerá una muestra de 400 casos mediante asignación proporcional por estrato.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# =========================================================
# CONFIGURACIÓN GENERAL
# =========================================================
SEED = 13094978
np.random.seed(SEED)

Path("../data/raw").mkdir(parents=True, exist_ok=True)
Path("../data/interim").mkdir(parents=True, exist_ok=True)
Path("../outputs").mkdir(parents=True, exist_ok=True)

RUTA_POBLACION = "../data/raw/poblacion_universitaria_simulada.csv"
RUTA_MUESTRA = "../data/raw/muestra_habitos_universitarios.csv"
RUTA_RESUMEN_MUESTREO = "../data/interim/resumen_muestreo_estratificado.csv"
RUTA_PROBABILIDADES = "../data/interim/probabilidades_eventos.csv"
RUTA_ARBOL = "../outputs/arbol_probabilidad_sueno_actividad.png"

N_POBLACION = 2000
N_MUESTRA = 400

# =========================================================
# 1) SIMULACIÓN DE LA POBLACIÓN
# =========================================================
id_estudiante = np.arange(10001, 10001 + N_POBLACION)

sexo = np.random.choice(
    ["Mujer", "Hombre", "Otro"],
    size=N_POBLACION,
    p=[0.50, 0.47, 0.03]
)

carrera_area = np.random.choice(
    ["Salud", "Ingeniería", "Ciencias Sociales", "Educación", "Administración"],
    size=N_POBLACION,
    p=[0.22, 0.24, 0.20, 0.14, 0.20]
)

anio_universitario = np.random.choice(
    [1, 2, 3, 4, 5],
    size=N_POBLACION,
    p=[0.24, 0.23, 0.20, 0.18, 0.15]
)

edad = np.random.randint(18, 30, size=N_POBLACION)

fumador = np.random.choice(["Sí", "No"], size=N_POBLACION, p=[0.14, 0.86])
consume_alcohol = np.random.choice(["Sí", "No"], size=N_POBLACION, p=[0.58, 0.42])

horas_estudio_dia = np.round(
    np.random.normal(2.4 + 0.35 * anio_universitario, 1.1, N_POBLACION),
    2
)
horas_estudio_dia = np.clip(horas_estudio_dia, 0.5, 10)

consumo_bebidas_energeticas_semana = np.random.poisson(
    0.8 + 0.18 * anio_universitario + 0.25 * (horas_estudio_dia > 5)
)
consumo_bebidas_energeticas_semana = np.clip(consumo_bebidas_energeticas_semana, 0, 10)

minutos_actividad_fisica_semana = np.round(
    np.random.gamma(shape=3.2, scale=52, size=N_POBLACION) - 10 * (anio_universitario >= 4),
    0
)
minutos_actividad_fisica_semana = np.clip(minutos_actividad_fisica_semana, 0, 600)

actividad_fisica_suficiente = np.where(
    minutos_actividad_fisica_semana >= 150,
    "Sí",
    "No"
)

actividad_bin = np.where(actividad_fisica_suficiente == "Sí", 1, 0)
fumador_bin = np.where(fumador == "Sí", 1, 0)
alcohol_bin = np.where(consume_alcohol == "Sí", 1, 0)

horas_sueno = np.round(
    np.random.normal(
        7.4
        - 0.22 * horas_estudio_dia
        - 0.16 * consumo_bebidas_energeticas_semana
        + 0.35 * actividad_bin
        - 0.10 * fumador_bin,
        0.75,
        N_POBLACION
    ),
    2
)
horas_sueno = np.clip(horas_sueno, 3.5, 10)

sueno_insuficiente = np.where(horas_sueno < 7, "Sí", "No")

nivel_estres_score = np.round(
    np.random.normal(
        5.0
        + 0.55 * horas_estudio_dia
        + 0.20 * consumo_bebidas_energeticas_semana
        - 0.25 * horas_sueno
        - 0.35 * actividad_bin,
        1.4,
        N_POBLACION
    )
)
nivel_estres_score = np.clip(nivel_estres_score, 1, 10).astype(int)

comidas_saludables_semana = np.random.poisson(
    7.5 + 0.8 * actividad_bin - 0.18 * nivel_estres_score
)
comidas_saludables_semana = np.clip(comidas_saludables_semana, 0, 21)

comidas_chatarra_semana = np.random.poisson(
    3.2 + 0.22 * nivel_estres_score + 0.18 * alcohol_bin - 0.25 * actividad_bin
)
comidas_chatarra_semana = np.clip(comidas_chatarra_semana, 0, 21)

calidad_sueno_score = np.round(
    np.random.normal(
        4.2
        + 0.65 * horas_sueno
        - 0.28 * nivel_estres_score
        - 0.12 * consumo_bebidas_energeticas_semana,
        1.0,
        N_POBLACION
    )
)
calidad_sueno_score = np.clip(calidad_sueno_score, 1, 10).astype(int)

imc = np.round(
    np.random.normal(
        23.8
        - 0.006 * minutos_actividad_fisica_semana
        + 0.16 * comidas_chatarra_semana
        - 0.09 * comidas_saludables_semana
        + 0.35 * fumador_bin,
        2.6,
        N_POBLACION
    ),
    2
)
imc = np.clip(imc, 16, 40)

imc_categoria = np.where(
    imc < 18.5, "Bajo peso",
    np.where(
        imc < 25, "Normal",
        np.where(imc < 30, "Sobrepeso", "Obesidad")
    )
)

salud_score = (
    4.6
    + 0.24 * horas_sueno
    + 0.006 * minutos_actividad_fisica_semana
    + 0.18 * comidas_saludables_semana
    - 0.16 * comidas_chatarra_semana
    - 0.28 * nivel_estres_score
)

autopercepcion_salud = np.where(
    salud_score < 4.5, "Mala",
    np.where(
        salud_score < 6.0, "Regular",
        np.where(salud_score < 7.5, "Buena", "Muy buena")
    )
)

df_poblacion = pd.DataFrame({
    "id_estudiante": id_estudiante,
    "edad": edad,
    "sexo": sexo,
    "carrera_area": carrera_area,
    "anio_universitario": anio_universitario,
    "horas_sueno": horas_sueno,
    "calidad_sueno_score": calidad_sueno_score,
    "comidas_saludables_semana": comidas_saludables_semana,
    "comidas_chatarra_semana": comidas_chatarra_semana,
    "minutos_actividad_fisica_semana": minutos_actividad_fisica_semana,
    "nivel_estres_score": nivel_estres_score,
    "horas_estudio_dia": horas_estudio_dia,
    "consumo_bebidas_energeticas_semana": consumo_bebidas_energeticas_semana,
    "imc": imc,
    "actividad_fisica_suficiente": actividad_fisica_suficiente,
    "sueno_insuficiente": sueno_insuficiente,
    "autopercepcion_salud": autopercepcion_salud,
    "fumador": fumador,
    "consume_alcohol": consume_alcohol,
    "imc_categoria": imc_categoria
})

df_poblacion.to_csv(RUTA_POBLACION, index=False)

# =========================================================
# 2) MUESTREO ESTRATIFICADO POR AÑO UNIVERSITARIO
# =========================================================
resumen_estratos = (
    df_poblacion["anio_universitario"]
    .value_counts()
    .sort_index()
    .rename_axis("anio_universitario")
    .reset_index(name="N_estrato")
)

resumen_estratos["proporcion"] = resumen_estratos["N_estrato"] / resumen_estratos["N_estrato"].sum()
resumen_estratos["n_muestra"] = np.round(resumen_estratos["proporcion"] * N_MUESTRA).astype(int)

# Ajuste para asegurar suma exacta = N_MUESTRA
diferencia = N_MUESTRA - resumen_estratos["n_muestra"].sum()
if diferencia != 0:
    idx_ajuste = resumen_estratos["n_muestra"].idxmax()
    resumen_estratos.loc[idx_ajuste, "n_muestra"] += diferencia

# Asegurar tipos enteros
resumen_estratos["anio_universitario"] = resumen_estratos["anio_universitario"].astype(int)
resumen_estratos["n_muestra"] = resumen_estratos["n_muestra"].astype(int)

muestras = []
for _, fila in resumen_estratos.iterrows():
    estrato = int(fila["anio_universitario"])
    n = int(fila["n_muestra"])

    df_estrato = df_poblacion[df_poblacion["anio_universitario"] == estrato]

    # salvaguarda por si alguna vez n supera el tamaño del estrato
    n = min(n, len(df_estrato))

    muestra_estrato = df_estrato.sample(
        n=n,
        random_state=SEED
    )
    muestras.append(muestra_estrato)

df_muestra = pd.concat(muestras, ignore_index=True)
df_muestra.to_csv(RUTA_MUESTRA, index=False)
resumen_estratos.to_csv(RUTA_RESUMEN_MUESTREO, index=False)

# =========================================================
# 3) DEFINICIÓN DE EVENTOS Y PROBABILIDADES BÁSICAS
# =========================================================
evento_A = df_muestra["actividad_fisica_suficiente"] == "Sí"
evento_B = df_muestra["sueno_insuficiente"] == "Sí"
evento_C = df_muestra["autopercepcion_salud"].isin(["Buena", "Muy buena"])

p_A = evento_A.mean()
p_B = evento_B.mean()
p_C = evento_C.mean()
p_A_inter_B = (evento_A & evento_B).mean()
p_A_union_B = (evento_A | evento_B).mean()
p_B_complemento = (~evento_B).mean()

probabilidades = pd.DataFrame([
    ["P(A)", "Actividad física suficiente", p_A],
    ["P(B)", "Sueño insuficiente", p_B],
    ["P(C)", "Autopercepción de salud favorable", p_C],
    ["P(A ∩ B)", "Actividad física suficiente y sueño insuficiente", p_A_inter_B],
    ["P(A ∪ B)", "Actividad física suficiente o sueño insuficiente", p_A_union_B],
    ["P(Bᶜ)", "Complemento de sueño insuficiente", p_B_complemento]
], columns=["evento", "descripcion", "probabilidad"])

probabilidades.to_csv(RUTA_PROBABILIDADES, index=False)

# =========================================================
# 4) ÁRBOL DE PROBABILIDAD SIMPLE: A y luego B
# =========================================================
p_no_A = 1 - p_A
p_B_dado_A = (evento_B[evento_A].mean()) if evento_A.sum() > 0 else 0
p_no_B_dado_A = 1 - p_B_dado_A

p_B_dado_no_A = (evento_B[~evento_A].mean()) if (~evento_A).sum() > 0 else 0
p_no_B_dado_no_A = 1 - p_B_dado_no_A

fig, ax = plt.subplots(figsize=(11, 6))
ax.axis("off")

# Nodos
ax.text(0.05, 0.50, "Inicio", fontsize=12, weight="bold")
ax.text(0.35, 0.75, "A: Actividad física suficiente", fontsize=11)
ax.text(0.35, 0.25, "Aᶜ: Actividad física insuficiente", fontsize=11)

ax.text(0.72, 0.90, "B: Sueño insuficiente", fontsize=10)
ax.text(0.72, 0.62, "Bᶜ: Sueño suficiente", fontsize=10)
ax.text(0.72, 0.38, "B: Sueño insuficiente", fontsize=10)
ax.text(0.72, 0.10, "Bᶜ: Sueño suficiente", fontsize=10)

# Ramas
ax.plot([0.10, 0.32], [0.50, 0.75], color="black")
ax.plot([0.10, 0.32], [0.50, 0.25], color="black")

ax.plot([0.52, 0.69], [0.75, 0.90], color="black")
ax.plot([0.52, 0.69], [0.75, 0.62], color="black")
ax.plot([0.52, 0.69], [0.25, 0.38], color="black")
ax.plot([0.52, 0.69], [0.25, 0.10], color="black")

# Etiquetas probabilidades
ax.text(0.18, 0.67, f"P(A) = {p_A:.3f}", fontsize=10)
ax.text(0.18, 0.35, f"P(Aᶜ) = {p_no_A:.3f}", fontsize=10)

ax.text(0.56, 0.84, f"P(B|A) = {p_B_dado_A:.3f}", fontsize=10)
ax.text(0.56, 0.69, f"P(Bᶜ|A) = {p_no_B_dado_A:.3f}", fontsize=10)
ax.text(0.56, 0.31, f"P(B|Aᶜ) = {p_B_dado_no_A:.3f}", fontsize=10)
ax.text(0.56, 0.16, f"P(Bᶜ|Aᶜ) = {p_no_B_dado_no_A:.3f}", fontsize=10)

plt.tight_layout()
plt.savefig(RUTA_ARBOL, dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# 5) RESULTADOS
# =========================================================
print("===== RESUMEN DEL MUESTREO ESTRATIFICADO =====")
display(resumen_estratos)

print("\n===== PRIMERAS 5 FILAS DE LA MUESTRA =====")
display(df_muestra.head())

print("\n===== PROBABILIDADES BÁSICAS =====")
display(probabilidades)

print("\n===== RESUMEN GENERAL =====")
print(f"Población simulada: {df_poblacion.shape[0]} registros")
print(f"Muestra simulada: {df_muestra.shape[0]} registros")
print(f"Archivo población guardado en: {RUTA_POBLACION}")
print(f"Archivo muestra guardado en: {RUTA_MUESTRA}")
print(f"Archivo resumen de muestreo guardado en: {RUTA_RESUMEN_MUESTREO}")
print(f"Archivo probabilidades guardado en: {RUTA_PROBABILIDADES}")
print(f"Árbol de probabilidad guardado en: {RUTA_ARBOL}")
```

## Lección 3 - Distribución de probabilidad

### Justificación de distribuciones seleccionadas

Para esta etapa se seleccionaron tres variables centrales del estudio y se asociaron a distribuciones de probabilidad coherentes con su naturaleza:

- **Sueño insuficiente**: se modelará mediante una distribución binomial, ya que representa un evento dicotómico (sí/no) en una cantidad fija de observaciones.
- **Consumo de bebidas energéticas por semana**: se modelará con una distribución de Poisson, dado que corresponde a un conteo de ocurrencias en un intervalo fijo de tiempo.
- **Horas de sueño**: se analizará como una variable continua con aproximación normal, debido a su naturaleza métrica y a la forma observada en su distribución muestral.

Estas distribuciones permiten representar adecuadamente distintos tipos de variables presentes en el estudio y sirven como base para el análisis inferencial posterior.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# =========================================================
# CONFIGURACIÓN
# =========================================================
RUTA_MUESTRA = "../data/raw/muestra_habitos_universitarios.csv"
RUTA_RESUMEN_DISTRIBUCIONES = "../data/interim/resumen_distribuciones.csv"

df_dist = pd.read_csv(RUTA_MUESTRA)

print("===== DIMENSIONES DE LA MUESTRA =====")
print(df_dist.shape)

# =========================================================
# 1) PARÁMETROS DE LAS DISTRIBUCIONES
# =========================================================
# Binomial: sueño insuficiente
p_sueno_insuf = (df_dist["sueno_insuficiente"] == "Sí").mean()

# Poisson: bebidas energéticas por semana
lambda_bebidas = df_dist["consumo_bebidas_energeticas_semana"].mean()

# Normal: horas de sueño
media_sueno = df_dist["horas_sueno"].mean()
std_sueno = df_dist["horas_sueno"].std(ddof=1)

print("\n===== PARÁMETROS ESTIMADOS =====")
print(f"p (sueño insuficiente) = {p_sueno_insuf:.4f}")
print(f"lambda (bebidas energéticas) = {lambda_bebidas:.4f}")
print(f"media (horas_sueno) = {media_sueno:.4f}")
print(f"desv_std (horas_sueno) = {std_sueno:.4f}")

# =========================================================
# 2) PROBABILIDADES CLAVE
# =========================================================
# BINOMIAL:
# Probabilidad de que exactamente 3 de 5 estudiantes tengan sueño insuficiente
prob_binomial = binom.pmf(k=3, n=5, p=p_sueno_insuf)

# POISSON:
# Probabilidad de que un estudiante consuma exactamente 2 bebidas energéticas por semana
prob_poisson = poisson.pmf(k=2, mu=lambda_bebidas)

# NORMAL:
# Probabilidad de que un estudiante duerma menos de 6 horas
prob_normal_menor_6 = norm.cdf(x=6, loc=media_sueno, scale=std_sueno)

# NORMAL:
# Probabilidad de que duerma entre 6 y 8 horas
prob_normal_6_8 = norm.cdf(x=8, loc=media_sueno, scale=std_sueno) - norm.cdf(
    x=6, loc=media_sueno, scale=std_sueno
)

resumen_distribuciones = pd.DataFrame([
    ["Binomial", "Sueño insuficiente", "P(X=3) con n=5", prob_binomial],
    ["Poisson", "Bebidas energéticas/semana", "P(X=2)", prob_poisson],
    ["Normal", "Horas de sueño", "P(X<6)", prob_normal_menor_6],
    ["Normal", "Horas de sueño", "P(6<X<8)", prob_normal_6_8]
], columns=["distribucion", "variable", "evento", "probabilidad"])

resumen_distribuciones.to_csv(RUTA_RESUMEN_DISTRIBUCIONES, index=False)

print("\n===== RESUMEN DE PROBABILIDADES =====")
display(resumen_distribuciones)

# =========================================================
# 3) GRÁFICOS DE DISTRIBUCIONES
# =========================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# -------------------------
# Binomial
# -------------------------
x_bin = np.arange(0, 6)
y_bin = binom.pmf(x_bin, n=5, p=p_sueno_insuf)

axes[0].bar(x_bin, y_bin)
axes[0].set_title("Distribución binomial\nSueño insuficiente en 5 estudiantes")
axes[0].set_xlabel("Número de estudiantes con sueño insuficiente")
axes[0].set_ylabel("Probabilidad")
axes[0].set_xticks(x_bin)

# -------------------------
# Poisson
# -------------------------
x_pois = np.arange(0, 11)
y_pois = poisson.pmf(x_pois, mu=lambda_bebidas)

axes[1].bar(x_pois, y_pois)
axes[1].set_title("Distribución de Poisson\nBebidas energéticas por semana")
axes[1].set_xlabel("Cantidad de bebidas energéticas")
axes[1].set_ylabel("Probabilidad")
axes[1].set_xticks(x_pois)

# -------------------------
# Normal
# -------------------------
x_norm = np.linspace(df_dist["horas_sueno"].min() - 1, df_dist["horas_sueno"].max() + 1, 300)
y_norm = norm.pdf(x_norm, loc=media_sueno, scale=std_sueno)

axes[2].hist(df_dist["horas_sueno"], bins=20, density=True, alpha=0.6)
axes[2].plot(x_norm, y_norm)
axes[2].set_title("Distribución normal aproximada\nHoras de sueño")
axes[2].set_xlabel("Horas de sueño")
axes[2].set_ylabel("Densidad")

plt.tight_layout()
plt.savefig("../outputs/distribuciones_probabilidad_modulo5.png", dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# 4) RESUMEN FINAL
# =========================================================
print("\n===== INTERPRETACIÓN INICIAL =====")
print(
    f"La probabilidad de que exactamente 3 de 5 estudiantes presenten sueño insuficiente es {prob_binomial:.4f}."
)
print(
    f"La probabilidad de que un estudiante consuma exactamente 2 bebidas energéticas por semana es {prob_poisson:.4f}."
)
print(
    f"La probabilidad estimada de que un estudiante duerma menos de 6 horas es {prob_normal_menor_6:.4f}."
)
print(
    f"La probabilidad estimada de que un estudiante duerma entre 6 y 8 horas es {prob_normal_6_8:.4f}."
)

print(f"\nArchivo de resumen guardado en: {RUTA_RESUMEN_DISTRIBUCIONES}")
print("Gráfico guardado en: ../outputs/distribuciones_probabilidad_modulo5.png")
```

### Hallazgos de la Lección 3

En esta etapa se identificaron distribuciones de probabilidad adecuadas para distintas variables del estudio, considerando su naturaleza discreta o continua. La variable sueño insuficiente fue modelada mediante una distribución binomial, el consumo de bebidas energéticas mediante una distribución de Poisson y las horas de sueño mediante una aproximación normal. A partir de estas distribuciones se calcularon probabilidades específicas de eventos relevantes para el problema de investigación, lo que permitió traducir el comportamiento observado en la muestra a un lenguaje probabilístico formal y científicamente justificable.

## Lección 4 - Distribución muestral y Teorema del Límite Central

### Objetivo de esta etapa

En esta etapa se analizará el comportamiento de la distribución muestral de la media utilizando la variable **horas de sueño**. Para ello, se generarán múltiples muestras aleatorias de distinto tamaño a partir de la población simulada, con el fin de observar empíricamente cómo se distribuyen las medias muestrales y cómo cambia su dispersión al aumentar el tamaño de la muestra.

Este procedimiento permite verificar de manera aplicada el **Teorema del Límite Central**, según el cual la distribución de la media muestral tiende a aproximarse a una distribución normal cuando el tamaño muestral es suficientemente grande, incluso si la distribución original de la variable no es perfectamente normal.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# =========================================================
# CONFIGURACIÓN
# =========================================================
RUTA_POBLACION = "../data/raw/poblacion_universitaria_simulada.csv"
RUTA_TLC_RESUMEN = "../data/interim/resumen_tlc.csv"

Path("../data/interim").mkdir(parents=True, exist_ok=True)
Path("../outputs").mkdir(parents=True, exist_ok=True)

SEED = 13094978
np.random.seed(SEED)

df_tlc = pd.read_csv(RUTA_POBLACION)

# =========================================================
# 1) PARÁMETROS DE LA POBLACIÓN
# =========================================================
variable_objetivo = "horas_sueno"
poblacion = df_tlc[variable_objetivo].values

media_poblacional = poblacion.mean()
std_poblacional = poblacion.std(ddof=1)

print("===== PARÁMETROS POBLACIONALES =====")
print(f"Variable analizada: {variable_objetivo}")
print(f"Media poblacional: {media_poblacional:.4f}")
print(f"Desviación estándar poblacional: {std_poblacional:.4f}")

# =========================================================
# 2) FUNCIÓN PARA GENERAR DISTRIBUCIÓN MUESTRAL DE LA MEDIA
# =========================================================
def generar_medias_muestrales(poblacion, n, repeticiones=1000):
    medias = []
    for _ in range(repeticiones):
        muestra = np.random.choice(poblacion, size=n, replace=True)
        medias.append(np.mean(muestra))
    return np.array(medias)

# Tamaños muestrales a comparar
tamanios_muestrales = [5, 30, 80]
resultados_medias = {}

for n in tamanios_muestrales:
    resultados_medias[n] = generar_medias_muestrales(poblacion, n=n, repeticiones=1000)

# =========================================================
# 3) RESUMEN DE LAS DISTRIBUCIONES MUESTRALES
# =========================================================
resumen_tlc = []

for n, medias in resultados_medias.items():
    resumen_tlc.append({
        "tamano_muestral": n,
        "media_de_las_medias": medias.mean(),
        "std_de_las_medias": medias.std(ddof=1),
        "error_estandar_teorico": std_poblacional / np.sqrt(n)
    })

df_resumen_tlc = pd.DataFrame(resumen_tlc)
df_resumen_tlc.to_csv(RUTA_TLC_RESUMEN, index=False)

print("\n===== RESUMEN DEL TLC =====")
display(df_resumen_tlc.round(4))

# =========================================================
# 4) COMPARACIÓN VISUAL: DISTRIBUCIÓN POBLACIONAL VS MUESTRALES
# =========================================================
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Distribución poblacional
axes[0, 0].hist(poblacion, bins=25, density=True, alpha=0.7)
axes[0, 0].axvline(media_poblacional, linestyle="--", label="Media poblacional")
axes[0, 0].set_title("Distribución poblacional de horas de sueño")
axes[0, 0].set_xlabel("Horas de sueño")
axes[0, 0].set_ylabel("Densidad")
axes[0, 0].legend()

# Distribuciones muestrales
posiciones = [(0, 1), (1, 0), (1, 1)]

for (n, medias), pos in zip(resultados_medias.items(), posiciones):
    ax = axes[pos]
    ax.hist(medias, bins=25, density=True, alpha=0.7)
    ax.axvline(media_poblacional, linestyle="--", label="Media poblacional")
    ax.set_title(f"Distribución muestral de la media (n={n})")
    ax.set_xlabel("Media muestral de horas de sueño")
    ax.set_ylabel("Densidad")
    ax.legend()

plt.tight_layout()
plt.savefig("../outputs/tlc_distribuciones_muestrales.png", dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# 5) COMPARACIÓN DE DISPERSIÓN SEGÚN TAMAÑO MUESTRAL
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(
    df_resumen_tlc["tamano_muestral"],
    df_resumen_tlc["std_de_las_medias"],
    marker="o",
    label="Desviación observada de las medias"
)
plt.plot(
    df_resumen_tlc["tamano_muestral"],
    df_resumen_tlc["error_estandar_teorico"],
    marker="s",
    label="Error estándar teórico"
)
plt.title("Dispersión de la media muestral según tamaño de muestra")
plt.xlabel("Tamaño muestral")
plt.ylabel("Desviación / Error estándar")
plt.xticks(tamanios_muestrales)
plt.legend()
plt.tight_layout()
plt.savefig("../outputs/tlc_dispersion_medias.png", dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# 6) INTERPRETACIÓN INICIAL
# =========================================================
print("\n===== INTERPRETACIÓN INICIAL =====")
for _, fila in df_resumen_tlc.iterrows():
    print(
        f"Para n = {int(fila['tamano_muestral'])}, la media de las medias fue {fila['media_de_las_medias']:.4f} "
        f"y la desviación de las medias fue {fila['std_de_las_medias']:.4f}."
    )

print("\nA medida que el tamaño muestral aumenta, la distribución de las medias se concentra más alrededor de la media poblacional.")
print("Esto constituye evidencia empírica a favor del Teorema del Límite Central.")

print(f"\nArchivo resumen guardado en: {RUTA_TLC_RESUMEN}")
print("Gráfico principal guardado en: ../outputs/tlc_distribuciones_muestrales.png")
print("Gráfico de dispersión guardado en: ../outputs/tlc_dispersion_medias.png")
```

### Hallazgos de la Lección 4

En esta etapa se generaron distribuciones muestrales de la media para la variable horas de sueño a partir de múltiples muestras aleatorias de distinto tamaño. Los resultados mostraron que, a medida que el tamaño muestral aumenta, las medias muestrales tienden a concentrarse en torno a la media poblacional y su dispersión disminuye. Esta evidencia empírica es consistente con el Teorema del Límite Central, ya que la distribución de las medias adquiere una forma progresivamente más estable y cercana a la normalidad conforme aumenta el tamaño de la muestra.

## Lección 5 - Inferencia e intervalos de confianza para la media

### Objetivo de esta etapa

En esta etapa se estimarán parámetros poblacionales mediante intervalos de confianza para la media, utilizando variables relevantes del estudio. Se trabajará con distintos niveles de confianza (90%, 95% y 99%) para analizar cómo cambia la precisión de la estimación. Además, se evaluará el efecto del tamaño muestral sobre el ancho del intervalo, con el fin de interpretar de manera rigurosa la relación entre muestra, incertidumbre y precisión estadística.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from pathlib import Path

# =========================================================
# CONFIGURACIÓN
# =========================================================
RUTA_MUESTRA = "../data/raw/muestra_habitos_universitarios.csv"
RUTA_IC = "../data/interim/intervalos_confianza_medias.csv"
RUTA_IC_TAMANOS = "../data/interim/impacto_tamano_muestral_ic.csv"

Path("../data/interim").mkdir(parents=True, exist_ok=True)
Path("../outputs").mkdir(parents=True, exist_ok=True)

SEED = 13094978
np.random.seed(SEED)

df_ic = pd.read_csv(RUTA_MUESTRA)

print("===== DIMENSIONES DE LA MUESTRA =====")
print(df_ic.shape)

# =========================================================
# 1) FUNCIÓN PARA INTERVALOS DE CONFIANZA PARA LA MEDIA
# =========================================================
def intervalo_confianza_media(serie, nivel_confianza=0.95):
    serie = pd.Series(serie).dropna()
    n = len(serie)
    media = serie.mean()
    s = serie.std(ddof=1)
    alpha = 1 - nivel_confianza
    t_critico = stats.t.ppf(1 - alpha / 2, df=n - 1)
    error = t_critico * (s / np.sqrt(n))
    li = media - error
    ls = media + error
    ancho = ls - li
    return {
        "n": n,
        "media_muestral": media,
        "desv_std_muestral": s,
        "nivel_confianza": nivel_confianza,
        "limite_inferior": li,
        "limite_superior": ls,
        "ancho_intervalo": ancho
    }

# =========================================================
# 2) IC PARA DOS VARIABLES Y TRES NIVELES DE CONFIANZA
# =========================================================
variables_objetivo = ["horas_sueno", "minutos_actividad_fisica_semana"]
niveles_confianza = [0.90, 0.95, 0.99]

resultados_ic = []

for variable in variables_objetivo:
    for nivel in niveles_confianza:
        resultado = intervalo_confianza_media(df_ic[variable], nivel)
        resultado["variable"] = variable
        resultados_ic.append(resultado)

df_intervalos = pd.DataFrame(resultados_ic)[[
    "variable", "n", "media_muestral", "desv_std_muestral",
    "nivel_confianza", "limite_inferior", "limite_superior", "ancho_intervalo"
]]

df_intervalos.to_csv(RUTA_IC, index=False)

print("\n===== INTERVALOS DE CONFIANZA PARA LA MEDIA =====")
display(df_intervalos.round(4))

# =========================================================
# 3) IMPACTO DEL TAMAÑO MUESTRAL EN EL ANCHO DEL INTERVALO
# =========================================================
# Usaremos horas_sueno y un nivel de confianza fijo del 95%
tamano_muestral_lista = [30, 60, 120, 240, 400]
impacto_tamano = []

serie_base = df_ic["horas_sueno"].dropna().sample(frac=1, random_state=SEED).reset_index(drop=True)

for n in tamano_muestral_lista:
    submuestra = serie_base.iloc[:n]
    resultado = intervalo_confianza_media(submuestra, 0.95)
    impacto_tamano.append({
        "tamano_muestral": n,
        "media_muestral": resultado["media_muestral"],
        "limite_inferior": resultado["limite_inferior"],
        "limite_superior": resultado["limite_superior"],
        "ancho_intervalo": resultado["ancho_intervalo"]
    })

df_impacto = pd.DataFrame(impacto_tamano)
df_impacto.to_csv(RUTA_IC_TAMANOS, index=False)

print("\n===== IMPACTO DEL TAMAÑO MUESTRAL EN EL IC (95%) =====")
display(df_impacto.round(4))

# =========================================================
# 4) GRÁFICO DEL ANCHO DEL INTERVALO SEGÚN n
# =========================================================
plt.figure(figsize=(10, 6))
plt.plot(df_impacto["tamano_muestral"], df_impacto["ancho_intervalo"], marker="o")
plt.title("Ancho del intervalo de confianza (95%) según tamaño muestral")
plt.xlabel("Tamaño muestral")
plt.ylabel("Ancho del intervalo")
plt.xticks(tamano_muestral_lista)
plt.tight_layout()
plt.savefig("../outputs/ancho_intervalo_confianza.png", dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# 5) VISUALIZACIÓN DE LOS INTERVALOS PARA HORAS DE SUEÑO
# =========================================================
df_horas = df_intervalos[df_intervalos["variable"] == "horas_sueno"].copy()
df_horas["nivel_pct"] = (df_horas["nivel_confianza"] * 100).astype(int).astype(str) + "%"

plt.figure(figsize=(10, 6))
plt.errorbar(
    x=df_horas["nivel_pct"],
    y=df_horas["media_muestral"],
    yerr=[
        df_horas["media_muestral"] - df_horas["limite_inferior"],
        df_horas["limite_superior"] - df_horas["media_muestral"]
    ],
    fmt="o",
    capsize=6
)
plt.title("Intervalos de confianza para la media de horas de sueño")
plt.xlabel("Nivel de confianza")
plt.ylabel("Horas de sueño")
plt.tight_layout()
plt.savefig("../outputs/ic_horas_sueno_niveles.png", dpi=300, bbox_inches="tight")
plt.show()

# =========================================================
# 6) INTERPRETACIÓN INICIAL
# =========================================================
print("\n===== INTERPRETACIÓN INICIAL =====")
for _, fila in df_intervalos.iterrows():
    print(
        f"Para la variable {fila['variable']} y un nivel de confianza del {int(fila['nivel_confianza']*100)}%, "
        f"la media estimada fue {fila['media_muestral']:.4f} y el intervalo de confianza fue "
        f"[{fila['limite_inferior']:.4f}, {fila['limite_superior']:.4f}]."
    )

print("\n===== CONCLUSIÓN SOBRE TAMAÑO MUESTRAL =====")
for _, fila in df_impacto.iterrows():
    print(
        f"Con n = {int(fila['tamano_muestral'])}, el ancho del intervalo fue {fila['ancho_intervalo']:.4f}."
    )

print("\nA medida que aumenta el tamaño muestral, el intervalo de confianza se vuelve más estrecho.")
print("Además, a mayor nivel de confianza, el intervalo tiende a ser más amplio.")

print(f"\nArchivo guardado en: {RUTA_IC}")
print(f"Archivo de impacto por tamaño muestral guardado en: {RUTA_IC_TAMANOS}")
print("Gráfico guardado en: ../outputs/ancho_intervalo_confianza.png")
print("Gráfico guardado en: ../outputs/ic_horas_sueno_niveles.png")
```

### Hallazgos de la Lección 5

En esta etapa se estimaron intervalos de confianza para la media de variables relevantes del estudio, utilizando distintos niveles de confianza. Los resultados mostraron que, a medida que aumenta el nivel de confianza, los intervalos se amplían, ya que se requiere mayor cobertura probabilística. Asimismo, al analizar el efecto del tamaño muestral sobre el ancho del intervalo, se observó que muestras más grandes producen estimaciones más precisas, reflejadas en intervalos más estrechos. Estos resultados son consistentes con la teoría inferencial y permiten interpretar con mayor rigurosidad la incertidumbre asociada a la estimación de parámetros poblacionales.

## Lección 6 - Test de significancia

### Objetivo de esta etapa

En esta etapa se aplicarán pruebas de hipótesis para evaluar afirmaciones sobre la población estudiada. En particular, se analizará si existen diferencias significativas en las horas de sueño promedio entre estudiantes con distinta suficiencia de actividad física, y además se evaluará si la proporción de estudiantes con sueño insuficiente supera un valor de referencia del 50%.

Las decisiones estadísticas se basarán en el valor-p y su comparación con un nivel de significancia α = 0.05.

```python
import pandas as pd
import numpy as np
from scipy import stats
from pathlib import Path

# =========================================================
# CONFIGURACIÓN
# =========================================================
RUTA_MUESTRA = "../data/raw/muestra_habitos_universitarios.csv"
RUTA_TESTS = "../data/interim/resultados_test_hipotesis.csv"

Path("../data/interim").mkdir(parents=True, exist_ok=True)

df_test = pd.read_csv(RUTA_MUESTRA)

alpha = 0.05

print("===== DIMENSIONES DE LA MUESTRA =====")
print(df_test.shape)

# =========================================================
# 1) TEST DE HIPÓTESIS PARA DIFERENCIA DE MEDIAS
# H0: μ1 = μ2
# H1: μ1 ≠ μ2
# =========================================================
grupo_activa = df_test.loc[
    df_test["actividad_fisica_suficiente"] == "Sí",
    "horas_sueno"
].dropna()

grupo_inactiva = df_test.loc[
    df_test["actividad_fisica_suficiente"] == "No",
    "horas_sueno"
].dropna()

t_stat, p_valor_t = stats.ttest_ind(
    grupo_activa,
    grupo_inactiva,
    equal_var=False  # Welch t-test
)

media_activa = grupo_activa.mean()
media_inactiva = grupo_inactiva.mean()
diferencia_medias = media_activa - media_inactiva

decision_t = "Rechazar H0" if p_valor_t < alpha else "No rechazar H0"

# =========================================================
# 2) TEST DE HIPÓTESIS PARA UNA PROPORCIÓN
# H0: p <= 0.50
# H1: p > 0.50
# =========================================================
n = len(df_test)
x = (df_test["sueno_insuficiente"] == "Sí").sum()
p_muestral = x / n
p0 = 0.50

z_stat = (p_muestral - p0) / np.sqrt((p0 * (1 - p0)) / n)
p_valor_z = 1 - stats.norm.cdf(z_stat)  # cola superior

decision_z = "Rechazar H0" if p_valor_z < alpha else "No rechazar H0"

# =========================================================
# 3) TABLA RESUMEN DE PRUEBAS
# =========================================================
resultados_tests = pd.DataFrame([
    {
        "prueba": "Diferencia de medias (t de Welch)",
        "hipotesis_nula": "No existe diferencia en horas_sueno entre grupos",
        "estadistico": t_stat,
        "p_valor": p_valor_t,
        "alpha": alpha,
        "decision": decision_t
    },
    {
        "prueba": "Proporción poblacional (z)",
        "hipotesis_nula": "La proporción de sueño insuficiente es <= 0.50",
        "estadistico": z_stat,
        "p_valor": p_valor_z,
        "alpha": alpha,
        "decision": decision_z
    }
])

resultados_tests.to_csv(RUTA_TESTS, index=False)

print("\n===== RESULTADOS DE LAS PRUEBAS DE HIPÓTESIS =====")
display(resultados_tests.round(6))

# =========================================================
# 4) DETALLE DE LA PRUEBA DE MEDIAS
# =========================================================
print("\n===== DETALLE TEST DE DIFERENCIA DE MEDIAS =====")
print(f"Media horas_sueno (actividad física suficiente): {media_activa:.4f}")
print(f"Media horas_sueno (actividad física insuficiente): {media_inactiva:.4f}")
print(f"Diferencia de medias: {diferencia_medias:.4f}")
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor-p: {p_valor_t:.6f}")
print(f"Decisión con α = {alpha}: {decision_t}")

# =========================================================
# 5) DETALLE DE LA PRUEBA DE PROPORCIÓN
# =========================================================
print("\n===== DETALLE TEST DE PROPORCIÓN =====")
print(f"Total de estudiantes con sueño insuficiente: {x}")
print(f"Tamaño muestral: {n}")
print(f"Proporción muestral: {p_muestral:.4f}")
print(f"Proporción bajo H0: {p0:.2f}")
print(f"Estadístico z: {z_stat:.4f}")
print(f"Valor-p: {p_valor_z:.6f}")
print(f"Decisión con α = {alpha}: {decision_z}")

# =========================================================
# 6) ERRORES TIPO I Y II EN CONTEXTO
# =========================================================
errores_contexto = pd.DataFrame([
    {
        "tipo_error": "Error Tipo I",
        "descripcion": "Rechazar una hipótesis nula verdadera",
        "aplicacion_en_el_estudio": "Concluir que existe diferencia en horas de sueño entre grupos o que la proporción de sueño insuficiente supera 50%, cuando en realidad no ocurre."
    },
    {
        "tipo_error": "Error Tipo II",
        "descripcion": "No rechazar una hipótesis nula falsa",
        "aplicacion_en_el_estudio": "No detectar una diferencia real en horas de sueño o no detectar que la proporción de sueño insuficiente sí supera 50%, cuando en realidad sí ocurre."
    }
])

print("\n===== ERRORES TIPO I Y II EN CONTEXTO =====")
display(errores_contexto)

# =========================================================
# 7) CONCLUSIÓN GENERAL AUTOMÁTICA
# =========================================================
print("\n===== CONCLUSIÓN GENERAL =====")
if p_valor_t < alpha:
    print(
        "La prueba de diferencia de medias entrega evidencia suficiente para afirmar que existen diferencias "
        "estadísticamente significativas en las horas de sueño promedio entre estudiantes con actividad física suficiente e insuficiente."
    )
else:
    print(
        "La prueba de diferencia de medias no entrega evidencia suficiente para afirmar diferencias estadísticamente significativas "
        "en las horas de sueño promedio entre los grupos analizados."
    )

if p_valor_z < alpha:
    print(
        "La prueba de proporción entrega evidencia suficiente para sostener que la proporción de estudiantes con sueño insuficiente "
        "es mayor al 50%."
    )
else:
    print(
        "La prueba de proporción no entrega evidencia suficiente para sostener que la proporción de estudiantes con sueño insuficiente "
        "sea mayor al 50%."
    )

print(f"\nArchivo guardado en: {RUTA_TESTS}")
```

Por otra, se analizó si la proporción de estudiantes con sueño insuficiente supera un umbral de referencia del 50% mediante una prueba para una proporción poblacional.
